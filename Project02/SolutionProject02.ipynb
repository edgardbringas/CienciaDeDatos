```python
# creamos funcion 
def fpaiscode(parm1):
   vpais = parm1;
   if vpais.upper() == 'USA':
      rcode = 1
   elif vpais.upper() == 'EUROPE':
      rcode = 2
   elif vpais.upper() == 'JAPAN':
      rcode = 3
   else: 
      rcode = 0
   return rcode
```

-	Generar 5 nuevas variables aplicando los siguientes criterios con las variables numéricas:

    ★	Log10(var1/var2)

    ★	Sqrt(var1)exp(var2)/200

    ★	Si var 1 > var 2 → 5, caso contrario 3

    ★	1/logn(var1/var2)*100

    ★	var2**2/var1

```python
# Copiamos el dataframe
df_rmx = df_rsnnn.copy()

# Log10(var1/var2)
df_rmx["columna_1"] = df_rmx.apply(lambda row: mt.log10(row["Desplazamiento"]/row["Aceleracion"]), axis=1)

# Sqrt(var1)exp(var2)/200
df_rmx["columna_2"] = df_rmx.apply(lambda row: mt.sqrt(row["Desplazamiento"])*mt.exp(row["Aceleracion"])/200, axis=1)

# Si var 1 > var 2 → 5, caso contrario 3
df_rmx["columna_3"] = df_rmx.apply(lambda row: 5 if row["Millas_Por_Galon"] > row["Aceleracion"] else 3, axis=1)

# 1/logn(var1/var2)*100
n = input("Ingrese el valor del logaritmo para el caso 1/logN(var1/var2)*100: ")
df_rmx["columna_4"] = df_rmx.apply(lambda row: 1/mt.log(int(n), row["Desplazamiento"]/row["Aceleracion"])*100, axis=1)

# var2**2/var1
df_rmx["columna_5"] = df_rmx.apply(lambda row: row["Desplazamiento"]**2/row["Aceleracion"], axis=1)

df_rmx
```

Generar 3 variables cualitativas ordinales con los siguientes criterios:

    ★	Si es mayor a la “media + 1*desviación estándar” → Alto

    ★	Si está entre “media - 1*desviación estándar” y “media + 1*desviación estándar” → Medio

    ★	Si es menor a la “media - 1*desviación estándar” → Bajo
Nota: Mostrar con las nuevas variables generadas. Elegir con libertad las variables a trabajar.

```python
# Copiamos el dataframe
df_rm = df_rsnnn.copy()

vmed = np.mean(df_rm["Millas_Por_Galon"])
vstd = np.std(df_rm["Millas_Por_Galon"])

# Rango Inicial
df_rm["RangoInicial"] = df_rm.apply(lambda row: (vmed - 1*vstd), axis=1)

# Si es mayor a la “media + 1*desviación estándar” → Alto
df_rm["columna_1"] = df_rm.apply(lambda row: "Alto" if row["Millas_Por_Galon"] >= (vmed + 1*vstd) else np.nan, axis=1)

# Si está entre “media - 1*desviación estándar” y “media + 1*desviación estándar” → Medio
df_rm["columna_2"] = df_rm.apply(lambda row: "Medio" if row["Millas_Por_Galon"] > (vmed - 1*vstd) and row["Millas_Por_Galon"] < (vmed + 1*vstd) else np.nan, axis=1)

# Si es menor a la “media - 1*desviación estándar” → Bajo
df_rm["columna_3"] = df_rm.apply(lambda row: "Bajo" if row["Millas_Por_Galon"] <= (vmed - 1*vstd) else np.nan, axis=1)

# Rango Final
df_rm["RangoFinal"] = df_rm.apply(lambda row: (vmed + 1*vstd), axis=1)

# mostrar las columnas que se estan evaluando
df_rm.loc[:, ["RangoInicial", "Millas_Por_Galon", "RangoFinal", "columna_1", "columna_2", "columna_3"]]

```
